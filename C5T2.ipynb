```python
#create histogram
plt.hist(credit['LIMIT_BAL'])
plt.show()
```

```python
#create histogram with 4 bins
plt.hist(credit['LIMIT_BAL'], bins = 4)
plt.show()
```

```python
#create line graph
plt.plot(credit['LIMIT_BAL'])

#Add title and axis names
plt.title('Line Graph Example')
plt.xlabel('Customers')
plt.ylabel('Limit Amount Balance')

#Display line plot
plt.show()
```

```python
#Build Scatter plot
x = credit['PAY_0']
y = credit['PAY_2']

plt.scatter(x,y)
#Add title and axis names
plt.title('Compare Payment Status for Sep & Aug 2005')
plt.xlabel('Payment Status for Sep. 2005')
plt.ylabel('Payment Status for Aug. 2005')
plt.show()
```

```python
#create box plots
header = credit.dtypes.index
print(header)
```

```python
A = credit['BILL_AMT1']
plt.boxplot(A,0,'gD')

plt.title('Scatter plot for bill amount in Sep. 2005')
plt.ylabel('Bill Amount in $')
plt.show()
```

```python
#build correlation
corrMat = credit.corr()
print(corrMat)
```

```python
covMat = credit.cov()
print(covMat)
```

```python
#histogram to show customers gender
p0 = credit['PAY_0']
p2 = credit['PAY_2']

fig, ax = plt.subplots()
ax.hist(p0, color='green', alpha=0.5)
ax.hist(p2, color='salmon', alpha=0.5)
ax.margins(0.05)
ax.set_ylim(bottom=0)
plt.title('Histogram with multiple variables')
plt.xlabel('Payment Status for Sep. 2005')
plt.ylabel('Payment Status for Aug. 2005')
plt.show()
```
